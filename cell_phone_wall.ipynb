```python
text_a = open("batterymythsdebunked.txt").read()
text_b = open("whatisyourphonesbatterytryingtotellyou.txt").read()
text_c = open("reasons.txt").read()
text_d = open("7waystokeepyourphonealive.txt").read()

import markovify

generator_a = markovify.Text(text_a)
generator_b = markovify.Text(text_b)
generator_c = markovify.Text(text_c)
generator_d = markovify.Text(text_d)
combo = markovify.combine([generator_a, generator_b, generator_c, generator_d], [0.25, 0.25, 0.25, 0.25])

for i in range(100):
    print(combo.make_short_sentence(120))
```
